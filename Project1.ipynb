### Step 1: Scraping SpaceWeatherLive

```python
# url to SpaceWeatherLive.com table
space_weather_live_url = "https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}


swl_response = requests.get(space_weather_live_url, headers = headers)

print("HTTP Code:", swl_response.status_code)

swl_text = swl_response.text
print("SpaceWeatherLive.com Raw Source:\n",swl_text)
```

```python
swl_soup = BeautifulSoup(swl_text, 'html.parser')
print(swl_soup.prettify())
```

```python
htmltable = swl_soup.find('table',{'class':'table table-striped'})
print(htmltable)
```

```python
swl_df = pd.read_html(str(htmltable))[0]
swl_df.head()
```
